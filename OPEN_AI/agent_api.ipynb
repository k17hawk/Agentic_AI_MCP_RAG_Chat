```python
import os
openai_api_key = os.environ.get("OPENAI_API_KEY")
```

```python
client = OpenAI(api_key=openai_api_key)
llm = client.chat.completions.create(
    model = 'gpt-3.5-turbo',
    messages=[{"role":'user',"content":"hello"}]
)
llm.choices[0].message.content
```
