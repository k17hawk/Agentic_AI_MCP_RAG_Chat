```python
from ollama import Client  

class User_history:
    def __init__(self, model: str, host: str):
        self.model = model
        self.host = host
        self.client = Client(host=self.host)
        self.messages = []
    
    def add_system_message(self, content: str):
        """Add a system message to the conversation history"""
        self.messages.append({'role': 'system', 'content': content})

    def add_user_message(self, content: str, stream: bool = True):
        """
        Add a user message and get assistant response
        
        Args:
            content: User message content
            stream: Whether to stream the response
            
        Returns:
            Full assistant response
        """
        self.messages.append({'role': 'user', 'content': content})
        
        try:
            response = self.client.chat(
                model=self.model,  # Make sure to include model parameter
                messages=self.messages,
                stream=stream
            )
            
            full_response = ''
            
            if stream:
                print("\nAssistant: ", end='', flush=True)
                for chunk in response:
                    if 'message' in chunk and 'content' in chunk['message']:
                        content = chunk['message']['content']
                        print(content, end='', flush=True)
                        full_response += content
                print()  # New line after streaming completes
            else:
                for chunk in response:
                    if 'message' in chunk and 'content' in chunk['message']:
                        full_response += chunk['message']['content']
            
            self.messages.append({'role': 'assistant', 'content': full_response})
            return full_response
            
        except Exception as e:
            print(f"Error during chat: {e}")
            return None
    
    def self_history(self):
        """Display conversation history"""
        if not self.messages:
            print("No messages in history")
            return
            
        for msg in self.messages:
            role = msg['role'].upper()
            content = msg['content']
            
            # Truncate long messages for display
            if len(content) > 100:
                display_content = content[:97] + "..."
            else:
                display_content = content
                
            print(f"{role}: {display_content}")
        print()
    
    def clear_history(self, flag: bool = False):
        """
        Clear conversation history
        
        Args:
            flag: If True, keep system messages; if False, clear all
        """
        if flag:
            self.messages = [msg for msg in self.messages if msg['role'] == 'system']
        else:
            self.messages = []


```

```python
user_history = User_history(model = 'gemma2:2b',host = 'http://localhost:11434')
user_history.add_system_message("You are an expert and helpful chatbot assistant")
user_history.add_user_message("Explain the theory of relativity?")
user_history.self_history()
```
