```python
from ollama import Client  

class User_history:
    def __init__(self, model: str, host: str):
        self.model = model
        self.host = host
        self.client = Client(host=self.host)
        self.messages = []
    
    def add_system_message(self, content: str):
        """Add a system message to the conversation history"""
        self.messages.append({'role': 'system', 'content': content})

    def add_user_message(self, content: str, stream: bool = True):
        """
        Add a user message and get assistant response
        
        Args:
            content: User message content
            stream: Whether to stream the response
            
        Returns:
            Full assistant response
        """
        self.messages.append({'role': 'user', 'content': content})
        
        try:
            response = self.client.chat(
                model=self.model, 
                messages=self.messages,
                stream=stream
            )
            
            full_response = ''
            
            if stream:
                print("\nAssistant: ", end='', flush=True)
                for chunk in response:
                    if 'message' in chunk and 'content' in chunk['message']:
                        content = chunk['message']['content']
                        print(content, end='', flush=True)
                        full_response += content
                print()  # New line after streaming completes
            else:
                for chunk in response:
                    if 'message' in chunk and 'content' in chunk['message']:
                        full_response += chunk['message']['content']
            
            self.messages.append({'role': 'assistant', 'content': full_response})
            return full_response
            
        except Exception as e:
            print(f"Error during chat: {e}")
            return None
    
    def self_history(self):
        """Display conversation history"""
        if not self.messages:
            print("No messages in history")
            return
            
        for msg in self.messages:
            role = msg['role'].upper()
            content = msg['content']
            
            # Truncate long messages for display
            if len(content) > 100:
                display_content = content[:97] + "..."
            else:
                display_content = content
                
            print(f"{role}: {display_content}")
        print()
    
    def clear_history(self, flag: bool = False):
        """
        Clear conversation history
        
        Args:
            flag: If True, keep system messages; if False, clear all
        """
        if flag:
            self.messages = [msg for msg in self.messages if msg['role'] == 'system']
        else:
            self.messages = []


```

```python
from ollama import Client
from datetime import datetime
import os
import json
import tiktoken 

class Chat_history:
    def __init__(self, model: str, host: str, max_context_tokens: int = 8192):
        self.model = model
        self.host = host
        self.max_context_tokens = max_context_tokens  # Set maximum allowed context
        self.warning_threshold = 0.85  # Warn at 85% of context limit
        self.messages = []
        self.conversation_id = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
        self.client = Client(host=self.host)
        
        # Initialize tokenizer (using cl100k_base which is used by many models)
        try:
            self.tokenizer = tiktoken.get_encoding("cl100k_base")
        except:
            # Fallback to a simple approximation if tiktoken fails
            self.tokenizer = None
            print("Warning: tiktoken not available, using approximate token counting")
        
        # System message
        system_message = {'role': 'system', 'content': 'You are a helpful chatbot assistant'}
        self.messages.append(system_message)
        
        # Track token count
        self.total_tokens = self._count_tokens_in_messages()
        print(f"Initialized with {self.total_tokens} tokens (System message)")
    
    def _count_tokens(self, text: str) -> int:
        """Count tokens in a text string"""
        if self.tokenizer:
            return len(self.tokenizer.encode(text))
        else:
            # Rough approximation: 1 token ‚âà 4 characters for English
            # This is very approximate and not accurate for all languages
            return len(text) // 4
    
    def _count_tokens_in_messages(self) -> int:
        """Count total tokens in all messages"""
        total = 0
        for msg in self.messages:
            # Add role tokens (approximate)
            total += 2  # Rough estimate for role formatting
            # Add content tokens
            total += self._count_tokens(msg['content'])
        return total
    
    def _check_token_limit(self, new_content: str = "") -> tuple:
        """
        Check if adding new content would exceed token limits
        Returns: (can_add, current_tokens, projected_tokens, percentage)
        """
        new_tokens = self._count_tokens(new_content) if new_content else 0
        
        # Account for assistant response (estimate - we don't know response length yet)
        estimated_response_tokens = self._count_tokens(new_content) * 2  # Rough estimate
        
        current_tokens = self.total_tokens
        projected_tokens = current_tokens + new_tokens + estimated_response_tokens
        percentage = (projected_tokens / self.max_context_tokens) * 100
        
        can_add = projected_tokens <= self.max_context_tokens
        
        return can_add, current_tokens, projected_tokens, percentage
    
    def _truncate_history(self, target_tokens: int):
        """Truncate conversation history to fit within token limit"""
        # Keep system message always
        system_messages = [msg for msg in self.messages if msg['role'] == 'system']
        non_system_messages = [msg for msg in self.messages if msg['role'] != 'system']
        
        # If we have no non-system messages, nothing to truncate
        if not non_system_messages:
            return
        
        # Calculate tokens used by system messages
        system_tokens = self._count_tokens_in_messages_for_list(system_messages)
        
        # Target tokens for non-system messages
        target_non_system = target_tokens - system_tokens
        
        # Remove oldest messages until we're under the limit
        truncated_messages = []
        current_tokens = 0
        
        # Keep newest messages (reverse to process from latest to oldest)
        for msg in reversed(non_system_messages):
            msg_tokens = self._count_tokens(msg['content']) + 2  # +2 for role
            if current_tokens + msg_tokens <= target_non_system:
                truncated_messages.insert(0, msg)  # Insert at beginning to maintain order
                current_tokens += msg_tokens
            else:
                # Add a warning message about truncation
                warning_msg = {
                    'role': 'system', 
                    'content': f'[System: Conversation history was truncated due to token limits. Some older messages were removed.]'
                }
                if warning_msg not in system_messages:
                    system_messages.append(warning_msg)
                break
        
        # Rebuild messages list
        self.messages = system_messages + truncated_messages
        self.total_tokens = self._count_tokens_in_messages()
        
        print(f"History truncated. New token count: {self.total_tokens}")
    
    def _count_tokens_in_messages_for_list(self, messages_list) -> int:
        """Count tokens for a specific list of messages"""
        total = 0
        for msg in messages_list:
            total += 2  # Role approximation
            total += self._count_tokens(msg['content'])
        return total
    
    def add_user_messages(self, content: str, stream: bool = True, auto_truncate: bool = True):
        """Add user message with token limit checking"""
        
        # Check token limits
        can_add, current_tokens, projected_tokens, percentage = self._check_token_limit(content)
        
        print(f"\n--- Token Usage ---")
        print(f"Current tokens: {current_tokens}/{self.max_context_tokens} ({current_tokens/self.max_context_tokens*100:.1f}%)")
        print(f"Projected tokens after response: {projected_tokens}/{self.max_context_tokens} ({percentage:.1f}%)")
        
        # Warning if approaching limit
        if percentage > self.warning_threshold * 100:
            print(f"‚ö†Ô∏è  WARNING: Approaching token limit ({percentage:.1f}%)")
        
        # Handle token limit exceeded
        if not can_add:
            if auto_truncate:
                print(f"‚ö†Ô∏è  Token limit would be exceeded. Attempting to truncate history...")
                # Target: keep enough space for new message and estimated response
                target_tokens = self.max_context_tokens - self._count_tokens(content) * 3
                self._truncate_history(target_tokens)
                
                # Recheck after truncation
                can_add, current_tokens, projected_tokens, percentage = self._check_token_limit(content)
                print(f"After truncation - Current: {current_tokens}, Projected: {projected_tokens} ({percentage:.1f}%)")
                
                if not can_add:
                    print(f"‚ùå ERROR: Even after truncation, token limit would be exceeded.")
                    print(f"Your input is too long for the context window.")
                    print(f"Please reduce your input length and try again.")
                    return None
            else:
                print(f"‚ùå ERROR: Token limit would be exceeded. Please reduce input length.")
                print(f"Current: {current_tokens}, Max: {self.max_context_tokens}")
                return None
        
        # Add user message
        user_message = {'role': 'user', 'content': content}
        self.messages.append(user_message)
        self.total_tokens += self._count_tokens(content) + 2  # +2 for role
        
        try:
            # Make API call with explicit context window
            response = self.client.chat(
                model=self.model,
                messages=self.messages,
                stream=stream,
                options={
                    "num_ctx": self.max_context_tokens,  # Set context window
                    "num_predict": 2048,  # Limit response length
                }
            )
            
            full_response = ""
            if stream:
                print("\nAssistant: ", end='', flush=True)
                for chunk in response:
                    if 'message' in chunk and 'content' in chunk['message']:
                        content_chunk = chunk['message']['content']
                        print(content_chunk, end='', flush=True)
                        full_response += content_chunk
                print()
            else:
                for chunk in response:
                    if 'message' in chunk and 'content' in chunk['message']:
                        full_response += chunk['message']['content']
            
            # Add assistant response to messages
            assistant_message = {'role': 'assistant', 'content': full_response}
            self.messages.append(assistant_message)
            
            # Update token count
            response_tokens = self._count_tokens(full_response)
            self.total_tokens += response_tokens + 2  # +2 for role
            
            # Save conversation
            self._save_chat_history()
            
            # Final token usage
            final_percentage = (self.total_tokens / self.max_context_tokens) * 100
            print(f"\n--- Final Token Usage: {self.total_tokens}/{self.max_context_tokens} ({final_percentage:.1f}%) ---")
            
            return full_response
            
        except Exception as e:
            print(f"Error during chat: {e}")
            # Remove the user message if there was an error
            self.messages.pop()
            self.total_tokens -= (self._count_tokens(content) + 2)
            return None
    
    def change_chat_role(self, system_message: str):
        """Change the system message"""
        # Remove old system messages
        self.messages = [msg for msg in self.messages if msg['role'] != 'system']
        # Add new system message
        self.messages.insert(0, {'role': 'system', 'content': system_message})
        # Recalculate token count
        self.total_tokens = self._count_tokens_in_messages()
        print(f"System message changed. New token count: {self.total_tokens}")
    
    def show_history(self, detailed: bool = False):
        """Display conversation history with token info"""
        if not self.messages:
            print("No messages in history")
            return
        
        print(f"\n{'='*60}")
        print(f"CONVERSATION HISTORY (Total Tokens: {self.total_tokens}/{self.max_context_tokens})")
        print(f"{'='*60}")
        
        for i, msg in enumerate(self.messages):
            role = msg['role'].upper()
            content = msg['content']
            
            # Count tokens for this message
            msg_tokens = self._count_tokens(content) + 2
            
            if detailed or len(content) <= 200:
                print(f"[{i}] {role} ({msg_tokens} tokens): {content}")
            else:
                display_content = content[:197] + "..."
                print(f"[{i}] {role} ({msg_tokens} tokens): {display_content}")
        
        print(f"{'='*60}\n")
    
    def clear_history(self, keep_system: bool = False):
        """Clear conversation history"""
        if keep_system:
            self.messages = [msg for msg in self.messages if msg['role'] == 'system']
        else:
            self.messages = []
        
        # Recalculate tokens
        self.total_tokens = self._count_tokens_in_messages()
        print(f"History cleared. Token count: {self.total_tokens}")
    
    def get_token_stats(self) -> dict:
        """Get detailed token statistics"""
        stats = {
            'total_tokens': self.total_tokens,
            'max_tokens': self.max_context_tokens,
            'percentage_used': (self.total_tokens / self.max_context_tokens) * 100,
            'messages_count': len(self.messages),
            'system_tokens': 0,
            'user_tokens': 0,
            'assistant_tokens': 0
        }
        
        for msg in self.messages:
            msg_tokens = self._count_tokens(msg['content']) + 2
            if msg['role'] == 'system':
                stats['system_tokens'] += msg_tokens
            elif msg['role'] == 'user':
                stats['user_tokens'] += msg_tokens
            elif msg['role'] == 'assistant':
                stats['assistant_tokens'] += msg_tokens
        
        return stats
    
    def _save_chat_history(self):
        """Save conversation to file"""
        file_path = 'conversations'
        if not os.path.exists(file_path):
            os.makedirs(file_path)
        
        file_name = os.path.join(file_path, f"{self.conversation_id}.json")
        
        # Get token stats for saving
        token_stats = self.get_token_stats()
        
        file_content = {
            'conversation_id': self.conversation_id,
            'model': self.model,
            'host': self.host,
            'max_context_tokens': self.max_context_tokens,
            'token_stats': token_stats,
            'message_count': len(self.messages),
            'messages': self.messages
        }
        
        with open(file_name, 'w') as f:
            json.dump(file_content, f, indent=4)
        
        return file_name


# Example usage
if __name__ == "__main__":
    # Initialize with model and host
    chat = Chat_history(
        model="gemma2:2b",  
        host="http://localhost:11434",
        max_context_tokens=8192  #
    )
    
```
